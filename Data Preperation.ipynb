```python
import os
import shutil
import glob
from tqdm import tqdm
```

```python
Raw_DIR= r'C:\Users\shaik\OneDrive\Desktop\Drowsiness detection sysyem4\mrlEyes_2018_01\mrlEyes_2018_01'
for dirpath, dirname, filenames in os.walk(Raw_DIR):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        if i.split('_')[4]=='0':
            shutil.copy(src=dirpath+'/'+i, dst=r'C:\Users\shaik\OneDrive\Desktop\Drowsiness detection sysyem4\Prepared_data\Closed_Eyes')
        
        elif i.split('_')[4]=='1':
            shutil.copy(src=dirpath+'/'+i, dst=r'C:\Users\shaik\OneDrive\Desktop\Drowsiness detection sysyem4\Prepared_data\Open_Eyes')
```

```python
# Cutting from Closed Eyes Main Folder To Test Closed Eyes Files Folder
src_dir = r'C:\Users\shaik\OneDrive\Desktop\Drowsiness detection sysyem4\Prepared_data\Open_Eyes'
dst_dir = r'C:\Users\shaik\OneDrive\Desktop\Drowsiness detection sysyem4\Prepared_data\test\Open_eyes'

file_counter = 0

for root, dirs, files in os.walk(src_dir):
    for file in files:
        if file.endswith('.png'):
            src_path = os.path.join(root, file)
            dst_path = os.path.join(dst_dir, file)
            shutil.move(src_path, dst_path)
            file_counter += 1
            if file_counter == 4200:
                break

print(f"Successfully moved {file_counter} PNG files to the destination directory.")
```
